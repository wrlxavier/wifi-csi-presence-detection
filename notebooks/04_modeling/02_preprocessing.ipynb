# 02 - Preprocessing (StandardScaler)

Fits a `StandardScaler` **on the training set only** and saves it to `models/scaler.pkl`. Subsequent notebooks load this scaler to transform val/test consistently.

```python
import sys
from pathlib import Path

_p = Path.cwd()
for ROOT in [_p, *_p.parents]:
    if (ROOT / "requirements.txt").exists():
        break
sys.path.insert(0, str(ROOT / "src"))

import joblib
from sklearn.preprocessing import StandardScaler

from preprocessing import load_feature_table, split_features_metadata

SPLITS_DIR = ROOT / "data" / "splits"
MODELS_DIR = ROOT / "models"
MODELS_DIR.mkdir(parents=True, exist_ok=True)
```

```python
train = load_feature_table(SPLITS_DIR / "train.parquet")
X_train, y_train, _ = split_features_metadata(train)

scaler = StandardScaler().fit(X_train)
joblib.dump(scaler, MODELS_DIR / "scaler.pkl")
print("scaler fit on", X_train.shape, "-> saved to", MODELS_DIR / "scaler.pkl")
```

```python
Xs = scaler.transform(X_train)
print("scaled train mean ~", round(float(Xs.mean()), 6))
print("scaled train std ~", round(float(Xs.std()), 6))
```
